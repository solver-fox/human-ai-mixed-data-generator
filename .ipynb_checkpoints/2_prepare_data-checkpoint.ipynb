# Prepare training data

Load **mixed_v2** and **sandwitch_v2** `.pt` chunks, window and augment, then merge with pure human/AI pickles and split into **train / validation / test** for word-level classification.

Each sample:
- **x** (`text`): document string (all sources: random **35–350 word** window)
- **y** (`label`): list of `0` (human) / `1` (AI) per word, aligned with `text.split()`

Mixed and sandwich sources are augmented **10×** with random windows that include at least one label boundary (0→1 or 1→0), excluding all-human or all-AI spans.

```python
import random
import sys
from pathlib import Path

import torch
from torch.utils.data import Dataset
import pickle

from generator.load_pt import list_chunks, load_chunk

PROJECT_DIR = Path.cwd()
sys.path.insert(0, str(PROJECT_DIR / "generator"))

# --- paths (chunk dirs from 1_data_generate.py) ---
MIXED_V2_DIR = PROJECT_DIR / "output" / "mixed_v2"
SANDWITCH_V2_DIR = PROJECT_DIR / "output" / "sandwitch_v2"
# Subfolders to scan under each mode dir (None = train, validation, test if present)
SOURCE_SPLITS = None  # e.g. ["train"] to only load train chunks

# --- train / val / test split ---
TRAIN_RATIO = 0.8
VAL_RATIO = 0.05
TEST_RATIO = 0.15
RANDOM_SEED = 42

# --- output ---
PREPARED_DIR = PROJECT_DIR / "output" / "prepared"

# --- window size (words) for all sample types ---
MIN_WINDOW_WORDS = 35
MAX_WINDOW_WORDS = 350

# --- mixed / sandwich windowing & augmentation ---
AUGMENT_COPIES = 10
AUGMENT_WORKERS = 0  # 0 = (CPU cores - 1); set 1 for single-process
AUGMENT_CHUNKSIZE = 128

# pure human / AI streaming targets
HUMAN_WINDOW_TARGET = 500_000
HUMAN_MAX_COPIES_PER_DOC = 8  # windows tried per doc while streaming
HUMAN_PARQUET_BATCH_ROWS = 2048  # rows per read (low RAM)

AI_WINDOW_TARGET = 500_000
AI_MAX_COPIES_PER_DOC = 16  # ~36k AI rows → need many windows per doc
AI_PARQUET_BATCH_ROWS = 2048
AI_ROW_LABEL_VALUE = 1  # keep rows with label==1 in ai-text-detection-training

from generator.prepare_subsample import (
    augment_mixed_samples,
    stream_pure_word_windows_from_parquet,
)
```

```python
def labels_to_list(labels) -> list[int]:
    if isinstance(labels, torch.Tensor):
        return [int(x) for x in labels.tolist()]
    return [int(x) for x in labels]


def validate_sample(text: str, label: list[int]) -> None:
    words = text.split()
    if len(words) != len(label):
        raise ValueError(
            f"Word count ({len(words)}) != label count ({len(label)}); sample misaligned."
        )
    if label and any(x not in (0, 1) for x in label):
        raise ValueError("Labels must be 0 (human) or 1 (AI).")


def discover_source_splits(output_dir: Path, source_splits=None) -> list[str]:
    if source_splits is not None:
        return list(source_splits)
    if not output_dir.exists():
        return []
    return sorted(
        d.name
        for d in output_dir.iterdir()
        if d.is_dir() and any(d.glob("*.pt"))
    )


def load_samples_from_dir(
    output_dir: Path, source_mode: str, source_splits=None
) -> list[dict]:
    """Load all samples from every .pt chunk under output_dir/<split>/."""
    if not output_dir.exists():
        print(f"Warning: missing directory {output_dir}")
        return []

    splits = discover_source_splits(output_dir, source_splits)
    if not splits:
        print(f"Warning: no .pt chunks under {output_dir}")
        return []

    samples = []
    n_chunks = 0
    for source_split in splits:
        paths = list_chunks(output_dir, source_split)
        n_chunks += len(paths)
        for path in paths:
            data = load_chunk(path)
            mode = data.get("generation_mode", source_mode)
            for i in range(len(data["texts"])):
                text = data["texts"][i]
                label = labels_to_list(data["labels"][i])
                try:
                    validate_sample(text, label)
                except ValueError:
                    print(f"Error: {path} {i}")
                    continue
                samples.append(
                    {
                        "text": text,
                        "label": label,
                        "source_mode": mode,
                        "minipile_index": int(data["indices"][i]),
                        "model": data["models"][i],
                        "chunk": path.name,
                    }
                )
    print(
        f"{source_mode}: {len(samples)} samples from {n_chunks} chunks "
        f"({', '.join(splits)}) in {output_dir}"
    )
    return samples


def merge_sources(*sample_lists: list[dict]) -> list[dict]:
    merged = []
    for batch in sample_lists:
        merged.extend(batch)
    return merged


def shuffle_samples(samples: list[dict], seed: int) -> list[dict]:
    rng = random.Random(seed)
    out = samples.copy()
    rng.shuffle(out)
    return out


def split_train_val_test(
    samples: list[dict],
    train_ratio: float,
    val_ratio: float,
    test_ratio: float,
) -> tuple[list[dict], list[dict], list[dict]]:
    total_ratio = train_ratio + val_ratio + test_ratio
    if abs(total_ratio - 1.0) > 1e-6:
        raise ValueError(f"Ratios must sum to 1.0, got {total_ratio}")

    n = len(samples)
    n_train = int(n * train_ratio)
    n_val = int(n * val_ratio)
    train = samples[:n_train]
    val = samples[n_train : n_train + n_val]
    test = samples[n_train + n_val :]
    return train, val, test
```

```python
import os
# create directory if it doesn't exist
os.makedirs("pickles", exist_ok=True)

# load all mixed samples and save as pickle
mixed_samples = load_samples_from_dir(MIXED_V2_DIR, "mixed_v2", SOURCE_SPLITS)
sandwich_samples = load_samples_from_dir(SANDWITCH_V2_DIR, "sandwitch_v2", SOURCE_SPLITS)
print("Mixed samples length: ", len(mixed_samples))
print("Sandwich samples length: ", len(sandwich_samples))

# Raw (full documents) — keep separate so augment can be re-run safely
mixed_raw = [{"text": d["text"], "label": d["label"]} for d in mixed_samples]
sandwich_raw = [{"text": d["text"], "label": d["label"]} for d in sandwich_samples]

with open("pickles/mixed_samples_raw.pkl", "wb") as f:
    pickle.dump(mixed_raw, f)

with open("pickles/sandwich_samples_raw.pkl", "wb") as f:
    pickle.dump(sandwich_raw, f)

print(f"Saved raw pickles: mixed={len(mixed_raw)}, sandwich={len(sandwich_raw)}")
```

```python
# Subsampling datasets (35 ~ 350 words, mixed windows only, N× augment)
import importlib
import generator.prepare_subsample as prepare_subsample

importlib.reload(prepare_subsample)  # pick up code fixes without kernel restart
from generator.prepare_subsample import augment_mixed_samples

RAW_MIXED = Path("pickles/mixed_samples_raw.pkl")
RAW_SANDWICH = Path("pickles/sandwich_samples_raw.pkl")

if not RAW_MIXED.exists() or not RAW_SANDWICH.exists():
    raise FileNotFoundError(
        "Run the previous cell first to create mixed_samples_raw.pkl and "
        "sandwich_samples_raw.pkl"
    )

with open(RAW_MIXED, "rb") as f:
    mixed_samples = pickle.load(f)

with open(RAW_SANDWICH, "rb") as f:
    sandwich_samples = pickle.load(f)

print(f"Before augment (raw) — mixed: {len(mixed_samples)}, sandwich: {len(sandwich_samples)}")

mixed_samples = augment_mixed_samples(
    mixed_samples,
    seed=RANDOM_SEED,
    copies_per_sample=AUGMENT_COPIES,
    min_words=MIN_WINDOW_WORDS,
    max_words=MAX_WINDOW_WORDS,
    desc="mixed augment",
    n_workers=AUGMENT_WORKERS,
    chunksize=AUGMENT_CHUNKSIZE,
)
sandwich_samples = augment_mixed_samples(
    sandwich_samples,
    seed=RANDOM_SEED + 1,
    copies_per_sample=AUGMENT_COPIES,
    min_words=MIN_WINDOW_WORDS,
    max_words=MAX_WINDOW_WORDS,
    desc="sandwich augment",
    n_workers=AUGMENT_WORKERS,
    chunksize=AUGMENT_CHUNKSIZE,
)

print(f"After augment — mixed: {len(mixed_samples)}, sandwich: {len(sandwich_samples)}")

with open("pickles/mixed_samples.pkl", "wb") as f:
    pickle.dump(mixed_samples, f)

with open("pickles/sandwich_samples.pkl", "wb") as f:
    pickle.dump(sandwich_samples, f)

```

```python
# Test: reload augmented pickles (pickles/mixed_samples.pkl, not *_raw.pkl)
# If this fails: run raw-load cell → augment cell, or: python regenerate_augmented_pickles.py
import pickle
from pathlib import Path

def has_label_boundary(labels: list[int]) -> bool:
    return any(labels[i] != labels[i + 1] for i in range(len(labels) - 1))


PICKLE_DIR = Path("pickles")

with open(PICKLE_DIR / "mixed_samples.pkl", "rb") as f:
    mixed_test = pickle.load(f)

with open(PICKLE_DIR / "sandwich_samples.pkl", "rb") as f:
    sandwich_test = pickle.load(f)

print(f"mixed: {len(mixed_test)} windows")
print(f"sandwich: {len(sandwich_test)} windows")


def check_subsampled(samples: list[dict], name: str, n_show: int = 3) -> None:
    word_lens = []
    bad_len = bad_align = bad_pure = bad_no_bound = 0

    for s in samples:
        text, labels = s["text"], s["label"]
        words = text.split()
        n_words = len(words)

        word_lens.append(n_words)
        if not (MIN_WINDOW_WORDS <= n_words <= MAX_WINDOW_WORDS):
            bad_len += 1
        if len(words) != len(labels):
            bad_align += 1
        if not (0 in labels and 1 in labels):
            bad_pure += 1
        if not has_label_boundary(labels):
            bad_no_bound += 1

    word_lens.sort()
    mid = len(word_lens) // 2
    print(f"\n--- {name} ---")
    print(f"  words: min={word_lens[0]}, median={word_lens[mid]}, max={word_lens[-1]} (target {MIN_WINDOW_WORDS}–{MAX_WINDOW_WORDS})")
    print(
        f"  checks: bad_len={bad_len}, bad_align={bad_align}, "
        f"bad_pure={bad_pure}, bad_no_bound={bad_no_bound}"
    )
    if bad_len or bad_align or bad_pure or bad_no_bound:
        raise AssertionError(
            f"{name} failed validation — re-run the augment cell to regenerate pickles "
            f"(stale pickles from before the subsample fix will fail here)."
        )

    for i in range(min(n_show, len(samples))):
        s = samples[i]
        text, labels = s["text"], s["label"]
        n_bounds = sum(1 for i in range(len(labels) - 1) if labels[i] != labels[i + 1])
        print(f"\n  [{i}] chars={len(text)} words={len(text.split())} boundaries={n_bounds}")
        print(f"      text: {text[:120]}{'...' if len(text) > 120 else ''}")
        print(f"      labels head: {labels[:20]}{'...' if len(labels) > 20 else ''}")


check_subsampled(mixed_test, "mixed")
check_subsampled(sandwich_test, "sandwich")
print("\nSubsampled pickles OK.")
```

```python
# load human data from pile dataset (from dataset/minipile/data)

def env_value(key, default=""):
    value = os.environ.get(key, default)
    return value.split("#", 1)[0].strip()

DATA_DIR = Path(env_value("MINIPILE_DATA_DIR", "dataset/minipile/data"))

files = sorted(Path(DATA_DIR).glob("train-*.parquet"))
if not files:
    raise FileNotFoundError(f"No train-*.parquet under {DATA_DIR}")

print(f"Streaming {len(files)} parquet file(s) → {HUMAN_WINDOW_TARGET} windows (batched, no multiprocessing)")

import importlib
import generator.prepare_subsample as prepare_subsample

importlib.reload(prepare_subsample)
from generator.prepare_subsample import stream_pure_word_windows_from_parquet

human_samples = stream_pure_word_windows_from_parquet(
    files,
    seed=RANDOM_SEED,
    target_count=HUMAN_WINDOW_TARGET,
    label=0,
    min_words=MIN_WINDOW_WORDS,
    max_words=MAX_WINDOW_WORDS,
    max_copies_per_doc=HUMAN_MAX_COPIES_PER_DOC,
    batch_rows=HUMAN_PARQUET_BATCH_ROWS,
    desc="human word window",
)

if len(human_samples) < HUMAN_WINDOW_TARGET:
    raise RuntimeError(
        f"Only {len(human_samples)} human windows; need {HUMAN_WINDOW_TARGET}. "
        "Add more parquet files under MINIPILE_DATA_DIR or raise HUMAN_MAX_COPIES_PER_DOC."
    )

with open("pickles/human_samples.pkl", "wb") as f:
    pickle.dump(human_samples, f)

print(f"Saved {len(human_samples)} human windows to pickles/human_samples.pkl")
```

```python
# load ai data from ai-text-detection-training (streamed, 35–350 words, 500k windows)

AI_DATA_DIR = Path("dataset/ai-text-detection-training")

ai_files = sorted(AI_DATA_DIR.glob("*.parquet"))
if not ai_files:
    raise FileNotFoundError(
        f"No parquet files found in {AI_DATA_DIR}. "
        "Check path and extracted dataset files."
    )

print(
    f"Streaming {len(ai_files)} parquet file(s) → {AI_WINDOW_TARGET} AI windows "
    f"(label=={AI_ROW_LABEL_VALUE} only, batched)"
)

import importlib
import generator.prepare_subsample as prepare_subsample

importlib.reload(prepare_subsample)
from generator.prepare_subsample import stream_pure_word_windows_from_parquet

ai_samples = stream_pure_word_windows_from_parquet(
    ai_files,
    seed=RANDOM_SEED + 2,
    target_count=AI_WINDOW_TARGET,
    label=1,
    min_words=MIN_WINDOW_WORDS,
    max_words=MAX_WINDOW_WORDS,
    max_copies_per_doc=AI_MAX_COPIES_PER_DOC,
    batch_rows=AI_PARQUET_BATCH_ROWS,
    desc="ai word window",
    text_column="text",
    row_label_column="label",
    row_label_value=AI_ROW_LABEL_VALUE,
)

if len(ai_samples) < AI_WINDOW_TARGET:
    raise RuntimeError(
        f"Only {len(ai_samples)} AI windows; need {AI_WINDOW_TARGET}. "
        "Raise AI_MAX_COPIES_PER_DOC or add more AI parquet data."
    )

with open("pickles/ai_texts.pkl", "wb") as f:
    pickle.dump(ai_samples, f)

print(f"Saved {len(ai_samples)} AI windows to pickles/ai_texts.pkl")
if ai_samples:
    print("Example:", ai_samples[0]["text"][:120], "...")
```

```python
# Load all pickles and inspect windows
from pathlib import Path

PICKLE_DIR = Path("pickles")

with open(PICKLE_DIR / "mixed_samples.pkl", "rb") as f:
    mixed_samples = pickle.load(f)
with open(PICKLE_DIR / "sandwich_samples.pkl", "rb") as f:
    sandwich_samples = pickle.load(f)
with open(PICKLE_DIR / "human_samples.pkl", "rb") as f:
    human_samples = pickle.load(f)
with open(PICKLE_DIR / "ai_texts.pkl", "rb") as f:
    ai_samples = pickle.load(f)

DATASETS = {
    "mixed": (mixed_samples, True),
    "sandwich": (sandwich_samples, True),
    "human": (human_samples, False),
    "ai": (ai_samples, False),
}


def word_count(text: str) -> int:
    return len(text.split())


def has_label_boundary(labels: list[int]) -> bool:
    return any(labels[i] != labels[i + 1] for i in range(len(labels) - 1))


def inspect_dataset(name, samples, need_boundary, n_show=2):
    print(f"\n{'=' * 60}")
    print(f"{name}: {len(samples):,} windows (words {MIN_WINDOW_WORDS}–{MAX_WINDOW_WORDS})")

    if not samples:
        print("  (empty)")
        return

    word_lens = []
    bad_align = bad_range = bad_pure = bad_no_bound = 0

    for s in samples:
        text, labels = s["text"], s["label"]
        n_words = word_count(text)
        word_lens.append(n_words)

        if n_words != len(labels):
            bad_align += 1
        if not (MIN_WINDOW_WORDS <= n_words <= MAX_WINDOW_WORDS):
            bad_range += 1
        if need_boundary:
            if not (0 in labels and 1 in labels):
                bad_pure += 1
            if not has_label_boundary(labels):
                bad_no_bound += 1
        else:
            expected = 0 if name == "human" else 1
            if labels and any(x != expected for x in labels):
                bad_pure += 1

    word_lens.sort()
    wm = word_lens[len(word_lens) // 2]
    print(f"  words/window: min={word_lens[0]}, median={wm}, max={word_lens[-1]}")
    print(
        f"  checks: bad_align={bad_align}, bad_word_range={bad_range}, "
        f"bad_label={bad_pure}"
        + (f", bad_no_bound={bad_no_bound}" if need_boundary else "")
    )

    for i in range(min(n_show, len(samples))):
        s = samples[i]
        text, labels = s["text"], s["label"]
        n_words = word_count(text)
        bounds = sum(1 for j in range(len(labels) - 1) if labels[j] != labels[j + 1])
        print(f"\n  --- sample [{i}] ---")
        print(f"  words={n_words}, chars={len(text)}, boundaries={bounds}")
        print(f"  text: {text[:200]}{'...' if len(text) > 200 else ''}")
        print(f"  labels (first 30): {labels[:30]}{'...' if len(labels) > 30 else ''}")


for name, (samples, need_bound) in DATASETS.items():
    inspect_dataset(name, samples, need_bound)

total = sum(len(v[0]) for v in DATASETS.values())
print(f"\n{'=' * 60}")
print(f"Total windows: {total:,}")
```

```python
# merge all samples
all_samples = merge_sources(mixed_samples, sandwich_samples, human_samples, ai_samples)

if not all_samples:
    raise RuntimeError(
        "No samples found. Generate data first, e.g.\n"
        "  python 1_data_generate.py --mode mixed_v2 ...\n"
        "  python 1_data_generate.py --mode sandwitch_v2 ..."
    )

all_samples = shuffle_samples(all_samples, RANDOM_SEED)
train, val, test = split_train_val_test(all_samples, TRAIN_RATIO, VAL_RATIO, TEST_RATIO)

print(f"Total merged: {len(all_samples)}")
print(f"train={len(train)}, val={len(val)}, test={len(test)}")

# save each split
for name, split_samples in [("train", train), ("validation", val), ("test", test)]:
    with open(f"pickles/{name}_samples.pkl", "wb") as f:
        pickle.dump(split_samples, f)
```
